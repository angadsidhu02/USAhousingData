```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('/content/USA_Housing.csv')
df=pd.DataFrame(data)
df.head()
```

```python
X=df.drop('Price', axis=1)
Y = df['Price']
X
```

```python
from sklearn.preprocessing import MinMaxScaler

selected_columns = ['Avg. Area Income', 'Avg. Area House Age'	, 'Avg. Area Number of Rooms'	, 'Avg. Area Number of Bedrooms'	, 'Area Population']

# Apply Min-Max Scaling
mx = MinMaxScaler()
X_scaled = mx.fit_transform(X)


# Create a new DataFrame with scaled values and original column names
X = pd.DataFrame(X_scaled, columns=selected_columns)
X
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

reg = LinearRegression()
reg.fit(X_train, y_train)
reg.score(X_test, y_test)


#kf = KFold(n_splits = 5)

#for train_index, test_index in kf.split(X):
    #X_train, X_test = X[train_index], X[test_index]
    #y_train, y_test = Y[train_index], Y[test_index]
```
